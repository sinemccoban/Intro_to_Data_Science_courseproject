Bu notlar, Matematik dersi ile ilgilidir:

G1 - birinci dönem notu (sayısal: 0'dan 20'ye kadar)

CSV dosyasını indiriyoruz.

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/AtaGn/Data/main/DataScienceClass/student-mat.csv',sep = ',')
df.head()
type(df)
```

Sınav notlarından ortalama sütunu yaratıldı ve ortalaması 10dan yüksek olanların dersi geçme durumu olarak son sütuna eklendi.

```python
df['Avg'] = (df['G1'] + df['G2'] + df['G3'])//3
df['Success'] = (df['Avg'] >= 10)
```

```python
df.head()
```

```python
df.shape
```

Sütunların bilgileri:

```python
df.info()
```

Sınav puanı sütunlarını ve ortalama puan sütununu siliyoruz.

```python
df = df.drop(['G1', 'G2', 'G3', 'Avg'], axis=1)
```

Null değere sahip veri kontrolü:

Satır ve sütunların isimlerini küçük harfle ve boşlukları alt çizgiyle dolduruyoruz. Bilgiyi işlerken sorun çıkmaması ve anlaşılır olması için.

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")
```

```python
string_columns = list(df.dtypes[df.dtypes=="object"].index)
for col in string_columns:
    df[col]=df[col].str.lower().str.replace(" ","_")
df.head()
```

Başarı bilgisini boolen'dan integer değerlere çeviriyoruz. Makine öğrenmesi uygularken gerekli olacak.

```python
df.success = df.success.astype(int)
df.head()
```

# Model Eğitimi

train_test_split komutuyla eğitim verisi ve test verisi ayrılıyor. Eğitim verisinin içerisinden modeli kontrol etmek için validation verisi ayırıyoruz.

```python
from sklearn.model_selection import train_test_split
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train, df_val = train_test_split(df_train_full, test_size=0.25, random_state=42)
y_train = df_train.success.values
y_val = df_val.success.values
del df_train["success"]
del df_val["success"]
```

Dictionary türüne çeviriyoruz:

```python
train_dict = df_train.to_dict(orient="records")
train_dict[:1]
```

Dictionary diziliminin ayarlanması:

```python
from sklearn.feature_extraction import DictVectorizer
dv = DictVectorizer(sparse = False)
dv.fit(train_dict)
X_train = dv.transform(train_dict)
X_train[0]
```

Özellik adlarının çıkartılması:

```python
dv.get_feature_names_out()
```

Makine öğrenmesi modelini yaratıyoruz. Eğitim verilerini veriyoruz.

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver="liblinear", random_state=42)
model.fit(X_train, y_train)
```

Kurulan modelin sonuçlarına bakıyoruz. İlk 5 öğrencinin geçme olasılıkları. 0 ve 1 in olma olasılıkları. 0 olasılığı kalma ihtimali, 1 olasılığı geçme ihtimalini temsil ediyor.

```python
val_dict = df_val.to_dict(orient="records")
X_val = dv.transform(val_dict)
y_pred = model.predict_proba(X_val)
y_pred[:5]
```

Makine öğrenmesi ile eğitilen modelin performansı:

```python
print("Validation verisinin performansı: ",model.score(X_val, y_val))
print("Eğitim verisinin performansı: ",model.score(X_train, y_train))
```

Parametrelerin değerleri:

```python
print("Bias: ",model.intercept_[0])
dict(zip(dv.get_feature_names_out(), model.coef_[0].round(3)))
```
